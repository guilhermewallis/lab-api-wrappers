```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="e5af6275806e4e8ebf14a5d71e202fad",
                                                           client_secret="a72673ed83bb437d83c801b760a066d5"))
```

```python
def get_playlist_tracks(username, playlist_id):

    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("spotify", "5KJDMJe9EJ7QRz8FG2MIpI")
tracks
```

```python
track_ids = [track['track']['uri'] for track in tracks]
track_ids
```

```python
start = 0
df = pd.DataFrame()

for stop in range(0, len(track_ids)+100, 100):
    if start != stop:
        print(start, stop)
        new_df = pd.json_normalize(sp.audio_features(track_ids[start:stop]))
        df = pd.concat([new_df, df])
        start = stop

df.reset_index()
```

```python
def get_artists_and_ids_from_playlist(playlist_id):

    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)

    artists = []
    artists_ids = []

    for track in tracks_from_playlist:
        artists_info = track['track']['artists']

        for artist_info in artists_info:
            artists.append(artist_info['name'])
            artists_ids.append(artist_info['id'])

    return list(set(artists)), list(set(artists_ids))

artists, artists_ids = get_artists_and_ids_from_playlist("5KJDMJe9EJ7QRz8FG2MIpI")
artists_ids
```

```python
df_artists = pd.DataFrame(list(zip(artists, artists_ids)),
               columns =['Name', 'ID'])
df_artists
```
